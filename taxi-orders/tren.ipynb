```python
import pandas as pd
import nltk
from nltk.corpus import stopwords as nltk_stopwords

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('tweets_lemm_train.csv')
test = pd.read_csv('tweets_lemm_test.csv')
```

```python
train 
```

```python
test
```

# code

```python
corpus = train['lemm_text'].values.astype('U')

nltk.download('stopwords')
stopwords = set(nltk_stopwords.words('russian'))

count_tf_idf = TfidfVectorizer(stop_words=stopwords)

tf_idf = count_tf_idf.fit_transform(corpus)
```

```python
tf_idf
```

```python
model = LogisticRegression()

model.fit(tf_idf, train['positive'])
```

```python
corpus_test = test['lemm_text'].values.astype('U')

tf_idf_test = count_tf_idf.transform(corpus_test)
```

```python
test['positive'] = model.predict(tf_idf_test)
test
```

```python
test.to_csv(r'C')
```
