* Accuracy (–∞–Ω–≥–ª. ¬´–ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å¬ª). –û—Ç–Ω–æ—à–µ–Ω–∏–µ —á–∏—Å–ª–∞ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –∫ —Ä–∞–∑–º–µ—Ä—É —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏.
* Precision (–∞–Ω–≥–ª. ¬´—Ç–æ—á–Ω–æ—Å—Ç—å¬ª). –û–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –∫–∞–∫ –º–Ω–æ–≥–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –Ω–∞—à–ª–∞ –º–æ–¥–µ–ª—å, –ø–æ–∫–∞ –∏—Å–∫–∞–ª–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ. –ß–µ–º –±–æ–ª—å—à–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö, —Ç–µ–º –Ω–∏–∂–µ —Ç–æ—á–Ω–æ—Å—Ç—å.
* Recall (–∞–Ω–≥–ª. ¬´–ø–æ–ª–Ω–æ—Ç–∞¬ª). –í—ã—è–≤–ª—è–µ—Ç, –∫–∞–∫—É—é –¥–æ–ª—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –æ—Ç–≤–µ—Ç–æ–≤ –≤—ã–¥–µ–ª–∏–ª–∞ –º–æ–¥–µ–ª—å. –í–∞–∂–Ω–æ –ø–æ–Ω–∏–º–∞—Ç—å, –∫–∞–∫ —Ö–æ—Ä–æ—à–æ –º–æ–¥–µ–ª—å –∏—Ö –Ω–∞—Ö–æ–¥–∏—Ç.

–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ:

```python
# accuracy, precision and recall for DecisionTree

model = DecisionTreeClassifier(max_depth=3, random_state=123)
model.fit(features_train, target_train)
predictions_train = model.predict(features_train) 
result_train = accuracy_score(target_train, predictions_train)
print(f'Accuracy –º–æ–¥–µ–ª–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ: {result_train:.1%}')

predictions_valid = model.predict(features_valid) 
result_vaid = accuracy_score(target_valid, predictions_valid)
print(f'Accuracy –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ: {result_vaid:.1%}')

predictions = model.predict(features_test) 
accuracy_tree = accuracy_score(target_test, predictions)
print(f'Accuracy –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {accuracy_tree:.1%}')

recall_tree = recall_score(target_test, predictions)
print(f'–ü–æ–ª–Ω–æ—Ç–∞ –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {recall_tree:.1%}')

precision_tree = precision_score(target_test, predictions)
print(f'–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {precision_tree:.1%}')
```
