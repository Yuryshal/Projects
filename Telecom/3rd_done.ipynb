#### –ü–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä

–û–±—Ä–∞—Ç–∏–º –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã `living_area –∏ kitchen_area`. –ü—Ä–æ–ø—É—Å–∫–∏ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö –ª–æ–≥–∏—á–Ω–æ –∑–∞–ø–æ–ª–Ω–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤ —Ç–∞–∫–æ–π –∂–µ –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ –æ—Ç –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏ `total_area`, –∫–∞–∫ —ç—Ç–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ –≤ –æ–±—â–µ–π —Ç–∞–±–ª–∏—Ü–µ.

–í—ã–ø–æ–ª–Ω–∏–º –ø–æ–¥—Å—á–µ—Ç –∂–∏–ª–æ–π –∏ –ø–ª–æ—â–∞–¥–∏ –¥–ª—è –∫—É—Ö–Ω–∏ —á–µ—Ä–µ–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ `living_ratio –∏ kitchen_ratio`, –∫–∞–∫ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–æ–ø–æ—Ä—Ü–∏–π. –û–ø—Ä–µ–¥–µ–ª–∏–º –¥–ª—è —ç—Ç–∏—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

```python
living_ratio = df['living_area'] / df['total_area'] # divide overall area on living area
```

```python
coef_living = living_ratio.mean() # average value of living_ratio
```

```python
kitchen_ratio = df['kitchen_area'] / df['total_area']  # divide overall area on kitchen area
```

```python
coef_kitchen = kitchen_ratio.mean() # average value of kitchen_ratio
```

–ü—Ä–∏–º–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –∂–∏–ª–æ–π –∏ –ø–ª–æ—â–∞–¥–∏ –¥–ª—è –∫—É—Ö–Ω–∏.

```python
# filling NaN with ratio of total area
df['living_area'] = df['living_area'].fillna(df['total_area']*coef_living) 
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area']*coef_kitchen)
```

```python
def floor_cat(row):   # function to define category
    if row['floor'] == 1:
        floor_cat = '–ø–µ—Ä–≤—ã–π'
    elif row['floor'] == row['floors_total']:
        floor_cat = '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        floor_cat = '–¥—Ä—É–≥–æ–π'
    return floor_cat

df['floor_cat'] = df.apply(floor_cat, axis=1) # apply function

df.head() # top 5 strings
```

```python
# query of data for St.Peterburg
    
spb = df_filter.query('locality_name=="–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥"').copy()
```

–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü `km` c –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º –¥–∞–Ω–Ω—ã—Ö –∏–∑ –∫–æ–ª–æ–Ω–∫–∏ `city_centers_nearest` –¥–æ –∫–∏–ª–æ–º–µ—Ç—Ä–æ–≤ (—Ü–µ–ª—ã–µ —á–∏–∏—Å–ª–∞).

```python
# columns 'spb' with values of km

spb['km'] = spb['city_centers_nearest']/1000 
spb['km'] = spb['km'].astype(int)
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –∫–∏–ª–æ–º–µ—Ç—Ä–∞ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞ –¥–ª—è –≤—Å–µ—Ö –æ–±—ä—è–≤–ª–µ–Ω–∏–π. –†–∞–∑–¥–µ–ª–∏–º –∫–æ–ª–æ–Ω–∫—É `price_meter` –Ω–∞ `km`.  –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∑–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü `price_km` —Å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º –¥–æ —Ü–µ–ª–æ–≥–æ. –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é `price_km`, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ–¥–µ—Ç –¥–∞–Ω–Ω—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Å –ø–æ–ø—Ä–∞–≤–∫–æ–π, –µ—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ `km` —Ä–∞–≤–Ω–æ –Ω—É–ª—é, —Ç–æ –≤ —Å—Ç–æ–ª–±–µ—Ü –ø—Ä–æ—Å—Ç–æ –≤–Ω–æ—Å–∏–º —á–∏—Å–ª–∞ –∏–∑ `price_meter` –±–µ–∑ –¥–µ–ª–µ–Ω–∏—è, —Ç.–∫. –Ω–∞ –Ω–æ–ª—å –¥–µ–ª–µ–Ω–∏–µ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ.

```python
def price_km(row):   # function for calculation of price_km
    if row['km'] == 0:
        price_km = row['price_meter']
    else:
        price_km = row['price_meter']/row['km']
    return int(price_km)

spb['price_km'] = spb.apply(price_km, axis=1) # apply function

spb.head() # top 5 strings of spb
```

–í—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–æ—à–ª–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ. –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –¥–ª—è —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∑–∞ –∫–∞–∂–¥—ã–π –∫–∏–ª–æ–º–µ—Ç—Ä.

```python
aver_price_spb = spb.pivot_table(index = 'km', values= 'price_km') # pivot_table for km and average price

aver_price_spb.plot(style='o-', grid=True) # plot for average price_km and km
```
